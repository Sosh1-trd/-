```python
x＝〔1,4,5,9,10〕y＝〔9,7.6.4,2〕
これらの点群を直線ｙ＝ax+bで近似してください。そのとき表示されるグラフの直線は赤色、点は青色にしてください。
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reg1dim2(x,y):
   n=len(x)
   a=((np.dot(x,y)-y.sum()*x.sum()/n)/
      ((x**2).sum()-x.sum()**2/n))
   b=(y.sum()-a*x.sum())/n
   return a,b

x=np.array([1,4,5,9,10])
y=np.array([9,7,6,4,2])
a,b=reg1dim2(x,y)

plt.scatter(x,y,color="b")
xmax=x.max()
plt.plot([0,xmax],[b,a*xmax+b],color="r")
plt.show()
```
